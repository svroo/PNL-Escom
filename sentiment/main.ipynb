```python
from nltk.sentiment import SentimentIntensityAnalyzer
import os
```

```python
path = 'C:\\Users\\hp\\Documents\\VSCode\\PNL-Escom\\Joder es cine\\corpusCriticasCine\\'

# Obtenemos una lista con todos los archivos en el corpus y los dividimos en x, y
name_files = os.listdir(path)
pos_file = list(filter(lambda x: x.endswith('.review.pos'), name_files))
xml_file = list(filter(lambda x: x.endswith('.xml'), name_files))


x, y = txt.obtener_y(path, pos_file, xml_file)       
print('x_file:', len(x),'\nY:', len(y))
```

```python
d = list()

# i = 0
for sentence in x:
    sid = SentimentIntensityAnalyzer()
    # print(sentence)
    # print(i)
    ss = sid.polarity_scores(sentence)
    d.append(ss)

    for k in sorted(ss):
        # print('{0}: {1}, '.format(k, ss[k]), end='')
        # print()
        if ss['pos'] > ss['neg'] :
            print("Polaridad positivs")
        elif ss['pos'] == ss['neg']:
            print("Neutral")
        elif ss['pos'] < ss['neg']:
            print("Negativo")
        
    # i+1
```
