## Riqueza lexica

Número de veces que aparece una palabra en un documento, calcularla para el text 8

```python
# Funciones para calcular la riqueza lexica y el porcentaje de repeticion de palabras en cada texto
def lexical_diversity(text):
    return len(text) / len(set(text))

def percentage(count, total):
    return 100 * count / total
```

```python
lexical_diversity(text8)
```

```python
lexical_diversity(text5)
```

```python
lexical_diversity(text1)
```
